# Bootstrapping
## An implementation of the bootstrapping method 

The example implemented below corresponds to the one from the paper "Bootstrapping" by Hagen Wey

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# collecting data
# ficticius blood pressure drop values after treatment for 80 patients

np.random.seed(42)
bloodPressureDrop = np.random.normal(loc=10, scale=4, size=80)
```

```python
# function to calculate the mean for a Bootstrap sample

def bootstrapMean(sample):
    return np.mean(sample)
```

```python
# application of the bootstrapping method

n_bootstrap = 1000
bootstrap_means = np.zeros(n_bootstrap)
```

```python
# botstrap samples and calculation of the bootstrap statistics

for i in range(n_bootstrap):
    bootstrap_sample = np.random.choice(bloodPressureDrop, size=len(bloodPressureDrop), replace=True)
    bootstrap_means[i] = bootstrapMean(bootstrap_sample)
```

```python
# confidence interval calculation

confidence_interval = np.percentile(bootstrap_means, [2.5, 97.5])
```

```python
# Results

print(f"95% confidence interval for the average drop in blood pressure: {confidence_interval[0]:.2f} till {confidence_interval[1]:.2f} mmHg")
```

```python
# Histogram of the bootstrap mean values
plt.hist(bootstrap_means, bins=30, edgecolor='k', alpha=0.7)
plt.axvline(confidence_interval[0], color='r', linestyle='--')
plt.axvline(confidence_interval[1], color='r', linestyle='--')
plt.title('Histogram of the bootstrap mean values')
plt.xlabel('Average drop in blood pressure (mmHg)')
plt.ylabel('Frequency')
plt.show()
```
